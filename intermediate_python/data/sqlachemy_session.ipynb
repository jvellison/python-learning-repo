```python
engine_gravity = create_engine("sqlite+pysqlite:////workspaces/python-learning-repo/intermediate_python/data/gravity.db")

connection_gravity = engine_gravity.connect()
inspection_gravity = inspect(connection_gravity)

inspection_gravity.get_table_names()
```

```python
metadata_903 = MetaData()

metadata_903.tables.keys() #empty

episodes = Table('episodes', metadata_903, autoload_with = engine_903)

print(episodes.name)
print(episodes.c)
print(episodes.c.keys())
print(repr(episodes))
```

```python
metadata_gravity = MetaData()

metadata_gravity.tables.keys() #empty

book = Table('book', metadata_gravity, autoload_with = engine_gravity) #choose a table name from db.get_table_names()

print(book.name)
print(book.c)
print(book.c.keys())
print(repr(book))
```

```python
#more complex query

header = Table('header', metadata_903, autoload_with = engine_903)

#df[col][(df[col] == value) | (df[col] == value2 )] #in pandas

with engine_903.connect() as con:
    #stat = select(header).where(header.columns.ETHNIC == "REFU") #return only rows where ETHNIC column = "REFU"
    stat = select(header).where(or_(header.columns.ETHNIC == "REFU",
                                    header.columns.ETHNIC == "NOBT")) #return only rows where ETHNIC column = "REFU"/"NOBT"
    print(stat)
    result_proxy = con.execute(stat)
    results = result_proxy.fetchmany(size=10)
    print(results)
```

```python
# select all of the book table where the publication_date is either 1999 or 2000

with engine_gravity.connect() as con:
    stat = select(book).where(or_(book.columns.book_id == 1,
                                  book.columns.book_id == 2)) # python
    print(stat) # print equivalent SQL code
    result_proxy = con.execute(stat)
    result = result_proxy.fetchmany(size=5)
    result2 = con.execute(stat).fetchmany(size=5) #in one line

print(result)
print(result2)
```
